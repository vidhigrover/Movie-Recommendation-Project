```python
movies_title=pd.read_csv('u.item',sep="\|",header=None)
```

```python
movies_titles=movies_title[[0,1]]
movies_titles.columns=["item_id","title"]
movies_titles.head(10)
```

```python
df=pd.merge(df,movies_titles,on="item_id")
```

```python
ratings=pd.DataFrame(df.groupby('title').mean()['rating'])
```

```python
ratings.head(10)
```

```python
ratings['num of ratings']=pd.DataFrame(df.groupby('title').count()['rating'])
```

```python
df.head(10)
```

```python
moviemat=df.pivot_table(index="user_id",columns="title",values="rating")
```

```python
moviemat.head(20)
```

```python
YoungFrankenstein_user_ratings=moviemat['Young Frankenstein (1974)']
```

```python
YoungFrankenstein_user_ratings.head(20)
```

```python
similar_to_YoungFrankenstein=moviemat.corrwith(YoungFrankenstein_user_ratings)
```

```python
similar_to_YoungFrankenstein
```

```python
corr_YoungFrankenstein=pd.DataFrame(similar_to_YoungFrankenstein,columns=['correlation']) 
```

```python
corr_YoungFrankenstein.dropna(inplace=True)
```

```python
corr_YoungFrankenstein
```

```python
corr_YoungFrankenstein.head(10) 
```

```python
corr_YoungFrankenstein=corr_YoungFrankenstein.join(ratings['num of ratings'])
```

```python
corr_YoungFrankenstein[corr_YoungFrankenstein['num of ratings']>100].sort_values('correlation',ascending=False)
```

```python
def predict_movies(movie_name):
    movie_user_ratings=moviemat[movie_name]
    similar_to_movie=moviemat.corrwith(movie_user_ratings)
    corr_movie=pd.DataFrame(similar_to_movie,columns=['correlation'])
    corr_movie.dropna(inplace=True)
    corr_movie=corr_movie.join(ratings['num of ratings'])
    
    predictions=corr_movie[corr_movie['num of ratings']>100].sort_values('correlation',ascending=False)
    
    return predictions
```

```python
predict_my_movie=predict_movies("Titanic (1997)")
```
